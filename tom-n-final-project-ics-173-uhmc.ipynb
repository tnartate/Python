This is my attempt at a final project with visualizations! I tried different types of visualizations that best fit the data that I was working with (grades referenced to parent cohabitation status). I used data taken from Kaggle.

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/kaggle/input/high-school-student-performance-and-demographics/student_math_clean.csv'
df = pd.read_csv(file_path)

fig, ax = plt.subplots(figsize=(10, 6))
ax.axis('off') 

table_data = df[['parent_status', 'grade_1', 'grade_2', 'final_grade']]
table = ax.table(cellText=table_data.values,
                 colLabels=table_data.columns,
                 cellLoc='center',
                 loc='center')

table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)  

plt.title('Raw Data Table')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/kaggle/input/high-school-student-performance-and-demographics/student_math_clean.csv'
df = pd.read_csv(file_path)

grouped_data = df.groupby('parent_status')[['grade_1', 'grade_2', 'final_grade']].mean()

fig, ax = plt.subplots(figsize=(10, 6))
bar_plot = grouped_data.plot(kind='bar', ax=ax, rot=0)
plt.title('Average Grades by Parent Cohabitation Status')
plt.xlabel('Parent Cohabitation Status')
plt.ylabel('Average Grade')
plt.legend(['Grade 1', 'Grade 2', 'Final Grade'])

ax.set_ylim(0, 20)

for p in bar_plot.patches:
    bar_plot.annotate(str(round(p.get_height(), 2)), (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/kaggle/input/high-school-student-performance-and-demographics/student_math_clean.csv'
df = pd.read_csv(file_path)

fig, ax = plt.subplots(figsize=(10, 6))

ax.scatter(df[df['parent_status'] == 'Apart']['parent_status'], 
           df[df['parent_status'] == 'Apart']['grade_1'], 
           label='Grade 1', alpha=0.7)

ax.scatter(df[df['parent_status'] == 'Apart']['parent_status'], 
           df[df['parent_status'] == 'Apart']['grade_2'], 
           label='Grade 2', alpha=0.7)

ax.scatter(df[df['parent_status'] == 'Apart']['parent_status'], 
           df[df['parent_status'] == 'Apart']['final_grade'], 
           label='Final Grade', alpha=0.7)

ax.scatter(df[df['parent_status'] == 'Living together']['parent_status'], 
           df[df['parent_status'] == 'Living together']['grade_1'], 
           label='Grade 1', alpha=0.7)

ax.scatter(df[df['parent_status'] == 'Living together']['parent_status'], 
           df[df['parent_status'] == 'Living together']['grade_2'], 
           label='Grade 2', alpha=0.7)

ax.scatter(df[df['parent_status'] == 'Living together']['parent_status'], 
           df[df['parent_status'] == 'Living together']['final_grade'], 
           label='Final Grade', alpha=0.7)

plt.title('Scatter Plot of Grades by Parent Cohabitation Status')
plt.xlabel('Parent Cohabitation Status')
plt.ylabel('Grade')
plt.legend()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '/kaggle/input/high-school-student-performance-and-demographics/student_math_clean.csv'
df = pd.read_csv(file_path)

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))

sns.swarmplot(x='parent_status', y='grade_1', data=df, color='blue', label='Grade 1', alpha=0.7)

sns.swarmplot(x='parent_status', y='grade_2', data=df, color='orange', label='Grade 2', alpha=0.7)

sns.swarmplot(x='parent_status', y='final_grade', data=df, color='green', label='Final Grade', alpha=0.7)

plt.title('Swarm Plot of Grades by Parent Cohabitation Status')
plt.xlabel('Parent Cohabitation Status')
plt.ylabel('Grade')

plt.legend(loc='upper right', bbox_to_anchor=(1.25, 1), title='Grades')

plt.show()

```
